## Customer Segmentation

For our customer segmentation we are going to use k-Means because it's an efficient, effective, and simple clustering algorithm.

```python
df['Total_Products'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds'] 
df.dropna(inplace=True)
cs = df[['Age', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Total_Products']]
cs
```

```python
from sklearn.cluster import KMeans
wcss = [] # minimizing within cluster sum of square
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(cs.iloc[:,1:])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(20,10))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.title('Elbow Method')
plt.tight_layout()
plt.savefig('Elbow_Method.png', bbox_inches = 'tight')
plt.show()
```

```python
km = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
clusters = km.fit_predict(cs.iloc[:,1:])

cs["label"] = clusters

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax.scatter(cs.Age[cs.label == 0], cs["Total_Products"][cs.label == 0], c='blue', s=60)
ax.scatter(cs.Age[cs.label == 1], cs["Total_Products"][cs.label == 1], c='red', s=60)
ax.scatter(cs.Age[cs.label == 2], cs["Total_Products"][cs.label == 2], c='green', s=60)
plt.xlabel("Age")
plt.ylabel("Total Products Purchased")
plt.title('Customer Segments')
plt.tight_layout()
plt.savefig('Customers_Segments.png', bbox_inches = 'tight')
plt.show()
```

The segments are divided into three clusters based on their age and the amount of products they bought.

```python
from yellowbrick.cluster import SilhouetteVisualizer
visualizer = SilhouetteVisualizer(km, size=(1080, 720))

visualizer.fit(cs.iloc[:,1:])    
visualizer.show(outpath="Silhouette.png") 
```

Silhouette analysis can be used to evaluate the density and separation between clusters. The score is calculated by averaging the silhouette coefficient for each sample, which is computed as the difference between the average intra-cluster distance and the mean nearest-cluster distance for each sample, normalized by the maximum value. This produces a score between -1 and +1, where scores near +1 indicate high separation and scores near -1 indicate that the samples may have been assigned to the wrong cluster.

## Predictive Model

To create our predictive model we are going to use Random Forest because it's a very powerful algorithm, easier to implement, interpret, and very efficient.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

features = df.drop(['ID','Year_Birth','Dt_Customer', 'Z_CostContact', 'Z_Revenue', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
                    'AcceptedCmp2','Response'], axis = 1) 
features = pd.get_dummies(features, columns=['Education', 'Marital_Status'])
labels = df['Response']

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42, n_jobs = -1, bootstrap=True)

rf.fit(train_features, train_labels)

predictions = rf.predict(test_features)
```

## Analyzing Model Perfomance

```python
from sklearn.metrics import accuracy_score
print('Accuracy:',round(accuracy_score(test_labels, predictions), 2),'%')
```

The accuracy answers the question:                                                                                                                                                       
How often is the classifier correct?                                                                                                                                                   
An accuracy of 89% is good, but it is important to have other ways to measure the performance of the algorithm.

```python
from yellowbrick.classifier import ClassificationReport, ROCAUC, ClassBalance,  ConfusionMatrix
from sklearn.naive_bayes import GaussianNB
bayes = GaussianNB()
visualizer = ROCAUC(rf, size=(1080, 720))

visualizer.fit(train_features, train_labels) 
visualizer.score(test_features, test_labels)  
g = visualizer.show(outpath="ROC_Curves.png")     
```

The receiver operating characteristic (ROC) curve it's created by plotting the true positive rate (TPR) (or recall) against the false positive rate (FPR).                               

An excellent model has AUC near to the 1 which means it has a good measure of separability.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_labels, predictions)
fig, ax = plt.subplots(figsize=(16,8)), plt.tight_layout()
cms = sns.heatmap(cm, annot=True, cmap="Blues").set_title('Confusion Matrix')
figure = cms.get_figure()    
figure.savefig('Confusion_Matrix', dpi=400)
```

* Out of the 475 actual instances of 0 (first row), the classifier predicted correctly 469 of them.
* Out of the 79 actual instances of 1 (second row), the classifier predicted correctly 22 of them.
* Out of all 554, the classifier predicted correctly 471 of them.

The result looks good. There is little difference between Test and Train result, indicating that our model does not overfit the data.
