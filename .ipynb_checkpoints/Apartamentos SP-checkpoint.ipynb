## OverView

O estado de São Paulo e, principalmente, sua capital possuem grande destaque no cenário econômico do Brasil, com uma grande presença de indústrias e uma enorme diversidade em atividades produtivas.                                                                                                                                                                   
Por esses motivos e vários outros, muitas pessoas tem o desejo de se mudar para a cidade, porém muitos desanimam quando veem o custo de vida elevado da capital, principalmente em questão de moradia. Pensando nisso resolvi fazer uma análise sobre os apartamentos em São Paulo, afim de entender melhor como é o mercado imobiliário em São Paulo. 
Os dados que irei usar são reais e foram retirados de um site de venda e aluguel de apartamentos.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('Apartamentos_sp.csv')
df.head(10)
```

## Bairros

Ao todo são 680 bairros no dataset que irei usar para análise, o que irá dificultar mas já era algo esperado.                                                                           
Portanto, para facilitar a comparação, irei usar os 15 bairros com mais ofertas.

```python
sns.set(font_scale=1.4)
df['Bairro'].value_counts().nlargest(15).sort_values(ascending=False).plot(kind='bar', color= '#85C1E9',figsize=(20, 10), rot=0)
plt.xlabel("Bairro", labelpad=14)
plt.xticks(rotation=45)
plt.ylabel("Apartamentos", labelpad=14)
plt.title("Contagem de apartamentos por bairro", y=1.02);
```

## O Principal Bairro

O bairro Santana possui mais de 700 ofertas de apartamento, sendo o principal bairro da nossa análise.                                                                                   
Para entender como estão distribuidos, vamos usar o bom e velho boxplot.

```python
data = df[['Bairro', 'Preço']]
most = df.query("Bairro == 'perdizes' | Bairro == 'santana' | Bairro == 'santa cecilia' | Bairro == 'tatuape' | Bairro == 'bela vista' | Bairro == 'pinheiros' | Bairro == 'bairra funda' | Bairro == 'higienopolis' | Bairro == 'consolacao' | Bairro == 'lauzane paulista' | Bairro == 'mooca' | Bairro == 'santa teresinha' | Bairro == 'cambuci' | Bairro == 'vila gomes cardim' | Bairro == 'vila regente feijo'")
sant = data.query("Bairro == 'santana'")
df['Bairro'].value_counts()
```

```python
sant.groupby('Bairro')['Preço'].plot(kind='box', color='#2874A6', figsize=(16, 20), rot=0)
#plt.xlabel("Apartamentos", labelpad=14)
#plt.xticks(rotation=90)
#plt.ylabel("Preço", labelpad=14)
plt.title("Preço dos Apartamentos", y=1.02);
```

## Preços
Como em todas as cidades o preço varia bastante, temos desde apartamentos bem humildes à condomínios luxuosos.                                                                           
Vamos dar uma olhada nos bairros com os menores e maiores preços, e nas médias e medianas.                                                                                               
Como em diversos bairros só existem uma oferta, para calcular os preços vou pegar somente os bairros com mais de 10 ofertas de apartamentos.

```python
df['Preço'].plot(kind='box', color='#2874A6', figsize=(16, 20), rot=0)
#plt.xlabel("Apartamentos", labelpad=14)
#plt.xticks(rotation=90)
#plt.ylabel("Preço", labelpad=14)
plt.title("Preço dos Apartamentos", y=1.02);
```

### Mínimo

```python
most.groupby('Bairro')['Preço'].min().nlargest(25).sort_values(ascending=True).plot(kind='barh', color='#5DADE2',figsize=(20, 15), rot=0)
plt.xlabel("Preço", labelpad=14)
#plt.xticks(rotation=90)
plt.ylabel("Bairro", labelpad=14)
plt.title("Preço Mínimo por Bairro", y=1.02);
```

### Mediana

```python
most.groupby('Bairro')['Preço'].median().nlargest(25).sort_values(ascending=True).plot(kind='barh', color='#5DADE2', figsize=(20, 15), rot=0)
plt.xlabel("Preço", labelpad=14)
#plt.xticks(rotation=90)
plt.ylabel("Bairro", labelpad=14)
plt.title("Mediana de Preço por Bairro", y=1.02);
```

### Média

```python
most.groupby('Bairro')['Preço'].mean().nlargest(25).sort_values(ascending=True).plot(kind='barh', color='#5DADE2', figsize=(20, 15), rot=0)
plt.xlabel("Preço", labelpad=14)
#plt.xticks(rotation=90)
plt.ylabel("Bairro", labelpad=14)
plt.title("Preço Médio", y=1.02);
```

### Máximo

```python
most.groupby('Bairro')['Preço'].max().nlargest(25).sort_values(ascending=True).plot(kind='barh', color='#5DADE2', figsize=(20, 15), rot=0)
plt.xlabel("Preço", labelpad=14)
#plt.xticks(rotation=90)
plt.ylabel("Bairro", labelpad=14)
plt.title("Preço Máximo", y=1.02);
```

### Correlação

```python
fig, ax = plt.subplots(figsize=(16,8))
plt.xticks(rotation=45)
sns.heatmap(df.corr(), cmap="Blues", annot=True)
```

## Conclusão
Podemos perceber que alguns bairros se destacam por possuirem preços muito atrativos e outros por serem muito caros, apesar de existirem 12 bairros com apartamentos por um preço inferior a 1000 Reais, a grande maioria fica acima dos 1500 nos principais bairros. Pesquisando um pouco mais a fundo, é possivel perceber que os apartamentos no centro são extremamente complicados, muitos deles são bem pequenos mas possuem um preço muito elevado, os apartamentos mais afastados são naturalmente bem mais baratos. Mudar de cidade ou até mesmo de apartamento é algo bem complexo que deve ser muito bem planejado, principalmente se for em São Paulo.
